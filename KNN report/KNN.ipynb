```python
import numpy as np
from tensorflow.keras.datasets import cifar10
```

```python
(x_train, y_train),(x_test, y_test)=cifar10.load_data()
```

```python
def cat_or_dog(label):
  return label==3 or label==5
```

```python
train_index=[]
test_index=[]
for i in range(len(y_train)):
  if cat_or_dog(y_train[i])==True:
    train_index.append(i)
for i in range(len(y_test)):
  if cat_or_dog(y_test[i])==True:
    test_index.append(i)
```

```python
x_train = x_train[train_index]
y_train = y_train[train_index]

x_test = x_test[test_index]
y_test = y_test[test_index]
```

```python
y_train = np.where(y_train == 3, 0, 1)
y_test = np.where(y_test == 3, 0, 1)
```

```python
import matplotlib.pyplot as plt
```

```python
x_train=x_train.astype("float")/255.0
x_test=x_test.astype("float")/255.0
```

```python
from sklearn.utils import shuffle
```

```python
x_train, y_train=shuffle(x_train, y_train)
```

```python
x_train=x_train.reshape(x_train.shape[0],-1)
x_test=x_test.reshape(x_test.shape[0],-1)
```

```python
from collections import Counter
```

```python
def eucliden_distance(X1,X2):
  return np.sqrt(np.sum((X1-X2)**2))
class K_Nearest_Neiborhood:
  def __init__(self,K=3):
    self.K=K
  def fit(self,features,labels):
    self.features=features
    self.labels=labels
  def predict(self,X):
    for feature in self.features:
      distances=eucliden_distance(X,feature)
    K_indices=np.argsort(distances)[:self.K]
    k_nearest_labels=[self.labels[i][0] for i in K_indices]
    most_common=Counter(k_nearest_labels).most_common(1)
    return most_common[0][0]
```

```python
model=K_Nearest_Neiborhood(11)
```

```python
model.fit(x_train,y_train)
```

```python
predictions=[]
for test in x_test:
  predictions.append(model.predict(test))
predictions=np.array(predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test, predictions)
accuracy
```
