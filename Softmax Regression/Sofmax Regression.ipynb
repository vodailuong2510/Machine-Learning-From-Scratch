```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from tensorflow.keras.datasets import cifar10
```

```python
(x_train, y_train), (x_test, y_test)=cifar10.load_data()
```

```python
x_train, x_test= x_train/255.0, x_test/255.0
```

```python
from sklearn.utils import shuffle
```

```python
x_train, y_train = shuffle(x_train, y_train)
```

```python
class_names = [
    'Airplane',
    'Automobile',
    'Bird',
    'Cat',
    'Deer',
    'Dog',
    'Frog',
    'Horse',
    'Ship',
    'Truck'
]
```

```python
plt.imshow(x_train[0])
plt.axis('off')
plt.title(class_names[int(y_train[0])])
plt.show()
```

```python
x_train = x_train.reshape(x_train.shape[0], -1)
x_test = x_test.reshape(x_test.shape[0], -1)
x_train=np.insert(x_train,0,1,axis=1)
x_test=np.insert(x_test,0,1,axis=1)
```

```python
from scipy import sparse
class Softmax:
  def __init__(self):
    self.w = None
  def fit(self, x, y, c, learning_rate=0.01, iters=100):
    np.random.seed(22520834)
    if self.w is None:
      self.w = np.random.rand(c, x.shape[1])
    self.y = self.convert_label(y,c)
    for iter in range(1, iters + 1):
      predictions = self.predict(x)
      accuracy = accuracy_score(y, predictions)
      if iter % 5 == 0:
        print(f"Completion: {iter/iters*100:.3f}%, accuracy: {accuracy:.4f}")
      mix_id = np.random.permutation(x.shape[0])
      for i in mix_id:
        xi = x[i].reshape(1, x.shape[1])
        yi = self.y[i].reshape(1, self.y.shape[1])
        ai = self.stable_softmax(xi.dot(self.w.T))
        self.w += learning_rate * (yi - ai).T.dot(xi)
  def predict(self, x):
    z = x.dot(self.w.T)
    a = self.stable_softmax(z)
    return np.argmax(a, axis=1)

  def stable_softmax(self, z):
    e_z = np.exp(z-np.max(z, axis = 1, keepdims= True))
    a = e_z / e_z.sum(axis=1, keepdims= True)
    return a

  def convert_label(self, y, c):
    y_one_hot = np.zeros((len(y), c))
    for i in range(len(y)):
      y_one_hot[i][int(y[i])]=1
    return y_one_hot
```

```python
model = Softmax()
print("Start training: ")
model.fit(x_train, y_train, c = len(class_names))
```
