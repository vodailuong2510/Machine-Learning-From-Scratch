```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10
from sklearn.metrics import accuracy_score
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
import matplotlib.pyplot as plt
from keras.datasets import cifar10
from sklearn.manifold import TSNE
import numpy as np

# Load the CIFAR-10 dataset


# Randomly sample a subset of the data (for example, 1000 samples)
sample_size = 1000
indices = np.random.choice(len(x_train), sample_size, replace=False)
x_train_subset = x_train[indices]
y_train_subset = y_train[indices]

# Flatten the images (convert each image to a 1D vector)
x_train_flat = x_train_subset.reshape(sample_size, -1)

# Reduce dimensionality using t-SNE
tsne = TSNE(n_components=2, random_state=0)
x_2d = tsne.fit_transform(x_train_flat)

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x_2d[:, 0], x_2d[:, 1], c=y_train_subset, cmap=plt.cm.tab10)
plt.title("t-SNE Visualization of CIFAR-10 Data (Subset)")
plt.colorbar()
plt.show()






```

```python
from sklearn.utils import shuffle
x_train, y_train = shuffle(x_train, y_train)
```

```python
class_names = [
    'Airplane',
    'Automobile',
    'Bird',
    'Cat',
    'Deer',
    'Dog',
    'Frog',
    'Horse',
    'Ship',
    'Truck'
]
```

```python
plt.imshow(x_train[0])
plt.axis('off')
plt.title(class_names[int(y_train[0])])
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Softmax
from tensorflow.keras.optimizers import SGD
```

```python
model = Sequential()
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
opt = SGD(learning_rate=0.01, momentum=0.9)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
y_one_hot = np.zeros((len(y_train), 10))
for i in range(len(y_train)):
  y_one_hot[i][int(y_train[i])]=1
y_train=y_one_hot
```

```python
model.fit(x_train, y_train, epochs=10)
```
